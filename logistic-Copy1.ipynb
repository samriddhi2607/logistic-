```python
from sklearn import datasets
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
traindata=pd.read_csv(r"C:\Users\TCS\Downloads\0000000000002429_training_titanic_x_y_train.csv",delimiter=",")
testdata=pd.read_csv(r"C:\Users\TCS\Downloads\0000000000002429_test_titanic_x_test.csv",delimiter=",")
print(traindata.isnull().sum())
```

```python
df=pd.DataFrame(traindata)
df.drop('Ticket',axis=1,inplace=True)
df.Age.fillna(df.Age.mean(),inplace=True)
df.Embarked.fillna('S',inplace=True)
def f1(s):
    if s=="S":
        return 0
    elif s=="C":
        return 1
    else:
        return 2

def f2(s):
    if s<=10:
        return 0
    
    elif s >10 and s<=20:
        return 1
    elif s >20 and s<=30:
        return 2
    else:
        
        return 3
    
    
    
def f(s):
    if s=="male":
        return 0
    else:
        return 1
    
    
def f3(s):
    if type(s)==float:
        return 0
    else:
        return 1
    
df["Gender"]=df.Sex.apply(f)
df.drop('Sex',axis=1,inplace=True)

df["namelen"]=df.Name.apply(len)
df.drop('Name',axis=1,inplace=True)

df["Embark"]=df.Embarked.apply(f1)
df.drop('Embarked',axis=1,inplace=True)

df["Fares"]=df.Fare.apply(f2)
df.drop('Fare',axis=1,inplace=True)

df["ncabin"]=df.Cabin.apply(f3)
df.drop('Cabin',axis=1,inplace=True)


df
```

```python
xtrain=df.drop("Survived",axis=1)
ytrain=df["Survived"]

print(xtrain.shape)
print(ytrain.shape)
```

```python
clf= LogisticRegression(solver='liblinear')
clf.fit(xtrain,ytrain)
```
